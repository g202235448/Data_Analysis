```python
# seaborn (relpot, displot, catplot)

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('C:/Users/user/Downloads/05_data1.xlsx')
df
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df6 = df.loc[:, ['gender', 'height', 'weight', 'waist', 'drinking', 'smoking']]
df6.loc[df['gender']==1, ['gender']]='M'
df6.loc[df['gender']==2, ['gender']]='F'
df6.loc[df['drinking']==0, ['drinking']]='Non-drinking'
df6.loc[df['drinking']==1, ['drinking']]='Drinking'
df6.loc[df['smoking']==1, ['smoking']]='Non-smoking'
df6.loc[df['smoking']==2, ['smoking']]='Quit-smoking'
df6.loc[df['smoking']==3, ['smoking']]='Smoking'
```

```python
drinking = df6.groupby(['gender', 'drinking'])['drinking'].count()
smoking = df6.groupby(['gender', 'smoking'])['smoking'].count()

drinking = drinking.to_frame(name='count')
smoking = smoking.to_frame(name='count')

drinking = drinking.reset_index()
smoking = smoking.reset_index()
```

```python
drinking
```

```python
smoking
```

```python
fig = plt.figure(figsize=(12,10))
fig.suptitle('2020 Health Check Drinking & Smoking Type Bar graph', fontweight='bold')
index1 = np.arange(4)
index2 = np.arange(6)

fig.add_subplot(2,1,1) # 2x1의 1번째를 의미

plt.bar(index1, drinking['count'])
plt.title('Drinking Type')
plt.ylabel('Count')
plt.xticks(index1, ['Non-drinking(M)', 'Drinking(M)', 'Non-drinking(F)', 'Drinking(F)'])

fig.add_subplot(2,1,2) # 2x1의 2번째를 의미

plt.bar(index2, smoking['count'])
plt.title('Smoking Type')
plt.ylabel('Count')
plt.xticks(index2, ['Non-smoking(M)', 'Quit-smoking(M)', 'Smoking(M)', 'Non-smoking(F)', 'Quit-smoking(F)', 'Smoking(F)'])

plt.show()
```

```python
fig = plt.figure(figsize=(12,12))

area1 = fig.add_subplot(2,1,1)
area2 = fig.add_subplot(2,1,2)

ax1 = sns.barplot(x='gender', y='count', hue='drinking', data=drinking, ax=area1)
ax2 = sns.barplot(x='gender', y='count', hue='smoking', data=smoking, ax=area2)

fig.suptitle('2020 Health Check Drinking & Smoking Type Seaborn Bar graph', fontweight='bold')
area1.set_title('Drinking Type')
area2.set_title('Smoking Type')

plt.show()
```

```python
# BMI (weight / height^2) 25 이상이면 비만, 23~25 사이면 과체중, 23 미만 정상
# (키는 m(미터) 기준)

df['BMI'] = (df['weight'] / ((df['height'] / 100) ** 2))
```

```python
df['BMI']
```

```python
# gender, height, weight, blood_sugar new dataframe으로 만들고, gender(M=1,F=2)로 변경, 
# blood_sugar >= 120 과 그렇지 않은 사람 구분 count해서 grouping 4개로

df4 = df.loc[:, ['gender', 'height', 'weight', 'blood_sugar']]
df4.loc[df['gender']==1, ['gender']]='M'
df4.loc[df['gender']==2, ['gender']]='F'

df4.loc[df['blood_sugar'] >= 120, ['blood_sugar']]='Over'
df4.loc[df['blood_sugar'] < 120, ['blood_sugar']]='Under'

df4.loc[df['BMI'] <= 23, ['BMI']]='Normal'
df4.loc[(df['BMI'] > 23) & (df['BMI'] < 25), ['BMI']]='Overweight'
df4.loc[df['BMI'] >= 25, ['BMI']]='Obesity'
```

```python
Blood_sugar = df4.groupby(['gender', 'blood_sugar'])['blood_sugar'].count()
BMI = df4.groupby(['gender', 'BMI'])['BMI'].count()

Blood_sugar = Blood_sugar.to_frame(name='count')
BMI = BMI.to_frame(name='count')

Blood_sugar = Blood_sugar.reset_index()
BMI = BMI.reset_index()
```

```python
Blood_sugar
```

```python
BMI
```

```python
fig = plt.figure(figsize=(12,10))
fig.suptitle('2020 Health Check Blood_sugar & BMI Type Bar graph', fontweight='bold')
index1 = np.arange(4)
index2 = np.arange(6)

fig.add_subplot(2,1,1) # 2x1의 1번째를 의미

plt.bar(index1, Blood_sugar['count'])
plt.title('Blood_sugar Type')
plt.ylabel('Count')
plt.xticks(index1, ['120 Over(M)', '120 Under(M)', '120 Over(F)', '120 Under(F)'])

fig.add_subplot(2,1,2) # 2x1의 2번째를 의미

plt.bar(index2, BMI['count'])
plt.title('BMI Type')
plt.ylabel('Count')
plt.xticks(index2, ['Normal(M)', 'Overweight(M)', 'Obesity(M)', 'Normal(F)', 'Overweight(F)', 'Obesity(F)'])

plt.show()
```

```python
fig = plt.figure(figsize=(12,12))

area1 = fig.add_subplot(2,1,1)
area2 = fig.add_subplot(2,1,2)

ax1 = sns.barplot(x='gender', y='count', hue='blood_sugar', data=Blood_sugar, ax=area1)
ax2 = sns.barplot(x='gender', y='count', hue='BMI', data=BMI, ax=area2)

fig.suptitle('2020 Health Check Blood_sugar & BMI Type Seaborn Bar graph', fontweight='bold')
area1.set_title('Blood_sugar Type')
area2.set_title('BMI Type')

plt.show()
```
