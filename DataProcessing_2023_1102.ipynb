```python
from collections import Counter
newdata['Genre'] = newdata['Genre'].str.strip()
genre_list = list()
for genres in newdata['Genre'].to_list():
    genres = genres.split(",  ")
    for gen in genres:
        genre_list.append(gen)       
genre_df = pd.DataFrame.from_dict(Counter(genre_list), orient = 'index').rename(columns = {0:'Count'})
genre_df.sort_values('Count',ascending = False, inplace = True)
genre_df.head()
```

```python
plt.figure(figsize=(25,10))
plt.plot(genre_df, color='r', marker='o')
plt.bar(genre_df.index, genre_df.Count)
plt.title('장르별 드라마 개수')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
# (도전과제) +5, 한국이나 미국 배경(OR)을 하고 있는 드라마의 방영요일 카운팅해서 df 만들고 그래프, 249, 250 참조, 선착순 3명 점수부여

newdata['KORUS'] = (newdata['KOR or USA']==True) # true인 값만 dataframe으로
KORUS = newdata[newdata['KORUS']]
KORUS
```

```python
from collections import Counter
KORUS['Aired On'] = KORUS['Aired On'].str.strip()
Aired_On_list = list()
for days in KORUS['Aired On'].to_list():
    days = days.split(", ")
    for day in days:
        Aired_On_list.append(day)
Aired_On_df = pd.DataFrame.from_dict(Counter(Aired_On_list), orient = 'index').rename(columns = {0:'Count'})
Aired_On_df.sort_values('Count', ascending = False, inplace = True)
Aired_On_df.head(10)
```

```python
plt.figure(figsize=(25,10))
plt.plot(Aired_On_df, color='r', marker='o')
plt.bar(Aired_On_df.index, Aired_On_df.Count)
plt.rcParams['font.size'] = 20
plt.title('Aired On Count based on KOR or US')
plt.xlabel('Aired On by KORUS')
plt.ylabel('Count')
plt.show()
```
