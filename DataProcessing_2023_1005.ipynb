```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('C:/Users/user/Downloads/04_data1.xlsx', index_col='area')
data
```

```python
plt.figure(figsize=(20,4))
plt.plot(data2020_good, color='b', marker='o', label='2020_good')
plt.plot(data2019_good, color='r', marker='s', label='2019_good')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2019 vs 2020 Fine Dust(pm10) good line graph')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('C:/Users/user/Downloads/04_data1.xlsx', index_col='area')
data2015_verybad = data['2015_verybad']
data2020_verybad = data['2020_verybad']

plt.figure(figsize=(20,4))
plt.bar(data2015_verybad.index, data2015_verybad, color='r', alpha=0.5, label='2020_verybad')
plt.bar(data2020_verybad.index, data2020_verybad, color='g', alpha=0.5, label='2015_verybad')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015 vs 2020 Fine Dust(pm10) verybad bar')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
index = np.arange(2)
print(index)
plt.figure(figsize=(20,4))
plt.bar(data2015_verybad.index, data2015_verybad, width=0.15, label='2015_verybad')
plt.bar(data2020_verybad.index, data2020_verybad, width=0.30, label='2020_verybad')
index = index + 0.15

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015 vs 2020 Fine Dust(pm10) verybad bar')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
index = np.arange(7)

plt.figure(figsize=(20,4))
data6 = data.loc['Seoul':'Ulsan', '2015_good':'2020_good']
for year in range(2015, 2021):
    year = str(year)+'_good'
    chartdata = data6[year]
    plt.bar(index, chartdata, width=0.15, label=year)
    index = index + 0.15

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.xticks(index-0.53, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2015 vs 2020 Fine Dust(pm10) verybad bar')
plt.legend()
plt.grid()
plt.show()
```

```python
iris = pd.read_csv('C:/Users/user/Downloads/04_data2.csv')
```

```python
PetalLength = iris['PetalLength']
PetalWidth = iris['PetalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris')
plt.grid()
plt.show()
```

```python
PetalLength = iris['PetalLength']
PetalWidth = iris['PetalWidth']
SepalLength = iris['SepalLength']
SepalWidth = iris['SepalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth, color='r', edgecolor='w',label ='PetalLength * PetalWidth')
plt.scatter(SepalLength, SepalWidth, color='b', edgecolor='w',label = 'SepalLength * SepalWidth')

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris')
plt.grid()
plt.show()
```

```python
# titanic scatter
import seaborn as sns

Age = titanic['Age']
Gender = titanic['Sex']

plt.figure(figsize=(10,4))
sns.scatterplot(x=Gender, y=Age, data=titanic)

plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Titanic')
plt.grid()
plt.show()
```

```python
versicolor_SepalWidth = iris.loc[iris.Species=='versicolor', ['Species', 'SepalWidth']]
setosa_SepalWidth = iris.loc[iris.Species=='setosa', ['Species', 'SepalWidth']]

plt.figure(figsize=(10,6))
plt.hist(versicolor_SepalWidth['SepalWidth'], bins=8, alpha=0.5, label='versicolor')
plt.hist(setosa_SepalWidth['SepalWidth'], bins=8, alpha=0.5, label='setosa')

plt.xlabel('Sepal')
plt.ylabel('Width')
plt.title('Iris Sepal Group Histogram')
plt.legend()
plt.grid()
plt.show()
```
