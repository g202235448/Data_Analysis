```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/Users/user/OneDrive/문서/카카오워크 받은 파일/test1026.csv')
df
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['dropoff_datetime'] = pd.to_datetime(df['dropoff_datetime'])
```

```python
df['weekday'] = df.pickup_datetime.dt.day_name()
df['month'] = df.pickup_datetime.dt.month
df['weekday_num'] = df.pickup_datetime.dt.weekday
df['pickup_hour'] = df.pickup_datetime.dt.hour
```

```python
# Vender에 따른 승객수

plt.figure(figsize=(10,6))
plt.title('Passenger per Hour using CountPlot Graph')
sns.countplot(x='passenger_count',data=df, hue='vendor_id')
```

```python
# 'pickup_hour', 'passenger_count', 'vendor_id', 'month'로만 구성된 새로운 newdf 구성

newdf = df.loc[:, ['pickup_hour', 'passenger_count', 'vendor_id', 'month']]
newdf
```

```python
#newdf의 히트맵 작성

plt.figure(figsize=(13,10))
plt.title('Heat Map Graph')

correlation_data3 = newdf.corr()
upp_mat = np.triu(correlation_data3)
sns.heatmap(correlation_data3, annot=True, cmap='YlGnBu', mask=upp_mat)
```

```python
# 승객이 0인 픽업 위도경도 로 구성된 zeropickup 생성

zeropickup = df.loc[df['passenger_count']==0, ['pickup_longitude', 'pickup_latitude']]
zeropickup.head()
```

```python
# 승객이 0인 드롭 위도경도 로 구성된 zeropickup 생성

zerodropoff = df.loc[df['passenger_count']==0, ['dropoff_longitude', 'dropoff_latitude']]
zerodropoff.head()
```

```python
# 픽업가 드롭에 따른 위도경도 산점도 작성

plt.figure(figsize=(10,4))
plt.scatter(zeropickup['pickup_longitude'], zeropickup['pickup_latitude'], color='r', edgecolor='w',label ='Pickup Location')
plt.scatter(zerodropoff['dropoff_longitude'], zerodropoff['dropoff_latitude'], color='b', edgecolor='w',label ='Dropoff Location')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Pickup & Dropoff Location')
plt.grid()
plt.legend()
plt.show()
```

```python
#'passenger_count', 'trip_duration', 'month', 'weekday', 'pickup_hour'로 구성된 data6 생성

data6 = df.loc[:, ['passenger_count', 'trip_duration', 'month', 'weekday', 'pickup_hour']]
data6
```

```python
data6.month.value_counts()
```

```python
# 1월과 5월에 승객수와 요일

#data6['month'] = data6['month'].replace({1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7: 'Sunday'})

hist1 = data6.loc[data6.month==1, ['passenger_count', 'weekday']]
hist2 = data6.loc[data6.month==5, ['passenger_count', 'weekday']]
```

```python
#hist1, hist2를 활용 히스토그램 작성

plt.figure(figsize=(10,6))
plt.hist(hist1['passenger_count'], bins=8, alpha=0.5, label='JAN')
plt.hist(hist2['passenger_count'], bins=8, alpha=0.5, label='MAY')

plt.xlabel('Weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger Count per WEEKDAY')
plt.legend()
plt.grid()
plt.show()
```

```python
#hist1, hist2를 활용 히스토그램 작성

plt.figure(figsize=(10,6))
plt.hist(hist1['weekday'], bins=8, alpha=0.5, label='JAN')
plt.hist(hist2['weekday'], bins=8, alpha=0.5, label='MAY')

plt.xlabel('Weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger Count per WEEKDAY')
plt.legend()
plt.grid()
plt.show()
```

```python
# 동일한 데이터로  KDE 히스토그램 작성

plt.figure(figsize=(10,6))
sns.histplot(hist1['passenger_count'], bins=8, alpha=0.5, label='JAN', color='g', kde=True)
sns.histplot(hist2['passenger_count'], bins=8, alpha=0.5, label='MAY', color='purple', kde=True)

plt.xlabel('weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger Count per WEEKDAY')
plt.legend()
plt.grid()
plt.show()
```
